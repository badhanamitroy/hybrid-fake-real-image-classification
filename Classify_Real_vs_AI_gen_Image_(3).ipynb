```python
# ==========================================
# 8. CHECK CORRUPTED IMAGES
# ==========================================
from PIL import Image

def find_corrupted_images(directory):
    corrupted = []
    print("\nScanning:", directory)

    for root, _, files in os.walk(directory):
        for f in files:
            if f.lower().endswith(('.jpg','.png','.jpeg','.bmp','.gif')):
                path = os.path.join(root, f)
                try:
                    with Image.open(path) as img:
                        img.verify()
                except:
                    corrupted.append(path)

    return corrupted


corrupted_train = find_corrupted_images(TRAIN_DIR)
corrupted_valid = find_corrupted_images(VALID_DIR)

print("\nTotal corrupted (train):", len(corrupted_train))
print("Total corrupted (valid):", len(corrupted_valid))
```

```python
# ==========================================
# 9. IMAGE GRID DISPLAY FUNCTION
# ==========================================
import numpy as np
import torchvision.utils as vutils
import matplotlib.pyplot as plt

def imshow(batch, title=None):
    grid = vutils.make_grid(batch)
    mean = np.array([0.485, 0.456, 0.406])
    std  = np.array([0.229, 0.224, 0.225])

    img = grid.numpy().transpose(1, 2, 0)
    img = std * img + mean
    img = np.clip(img, 0, 1)

    plt.imshow(img)
    if title: plt.title(title)
    plt.axis("off")
    plt.show()
```

# 11. Classical Feature Extraction (HOG + LBP + Sobel)

```python
# ======================================================
# 11. CLASSICAL FEATURE EXTRACTION FOR RANDOM FOREST
# ======================================================
import cv2
import numpy as np
from skimage.feature import hog, local_binary_pattern

def extract_features(img_path):
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    img = cv2.resize(img, (128, 128))

    # --- HOG ---
    hog_feat = hog(img, orientations=9, pixels_per_cell=(8, 8),
                   cells_per_block=(2, 2), block_norm='L2-Hys')

    # --- LBP ---
    lbp = local_binary_pattern(img, P=8, R=1, method='uniform')
    lbp_hist, _ = np.histogram(lbp.ravel(), bins=256, range=(0, 256))

    # --- Sobel (Edges) ---
    sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0)
    sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1)
    sobel_feat = np.hstack([sobelx.ravel(), sobely.ravel()])

    # Combine all features
    return np.hstack([hog_feat, lbp_hist, sobel_feat])
```

# 19. ROC CURVES (One-vs-Rest for Multi-class)

```python
# ======================================================
# 19. ROC CURVES (One-vs-Rest for Multi-class)
# ======================================================
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc
from itertools import cycle
import numpy as np # Ensure numpy is imported for np.eye

# Binarize the labels for ROC computation
# The original label_binarize with a 1D y_val returns a single column for binary classification.
# To get one-hot encoding for each class for "one-vs-rest" ROC, we need to manually create it.
y_val_numeric = np.array([train_dataset.class_to_idx[cls] for cls in y_val])
n_classes = len(train_dataset.classes)
y_val_bin = np.eye(n_classes)[y_val_numeric]

# Compute ROC curve and ROC area for each class
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_val_bin[:, i], y_pred_proba[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Plot ROC curves
fig, ax = plt.subplots(figsize=(12, 8))
colors = cycle(['blue', 'red', 'green', 'orange', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan'])

for i, color in zip(range(n_classes), colors):
    ax.plot(fpr[i], tpr[i], color=color, lw=2,
            label=f'{train_dataset.classes[i]} (AUC = {roc_auc[i]:.3f})')

ax.plot([0, 1], [0, 1], 'k--', lw=2, label='Random Guess')
ax.set_xlim([0.0, 1.0])
ax.set_ylim([0.0, 1.05])
ax.set_xlabel('False Positive Rate', fontsize=12)
ax.set_ylabel('True Positive Rate', fontsize=12)
ax.set_title('ROC Curves - Random Forest (One-vs-Rest)', fontsize=16)
ax.legend(loc="lower right", fontsize=10)
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

# 20. MACRO AND MICRO AVERAGE ROC

```python
# ======================================================
# 20. MACRO AND MICRO AVERAGE ROC
# ======================================================
import numpy as np # Ensure numpy is imported for np.interp

# Compute micro-average ROC curve and ROC area
fpr["micro"], tpr["micro"], _ = roc_curve(y_val_bin.ravel(), y_pred_proba.ravel())
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

# Compute macro-average ROC curve and ROC area
all_fpr = np.unique(np.concatenate([fpr[i] for i in range(n_classes)]))
mean_tpr = np.zeros_like(all_fpr)
for i in range(n_classes):
    mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])
mean_tpr /= n_classes

fpr["macro"] = all_fpr
tpr["macro"] = mean_tpr
roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])

# Plot average ROC curves
fig, ax = plt.subplots(figsize=(10, 8))
ax.plot(fpr["micro"], tpr["micro"],
        label=f'Micro-average (AUC = {roc_auc["micro"]:.3f})',
        color='deeppink', linestyle=':', linewidth=3)

ax.plot(fpr["macro"], tpr["macro"],
        label=f'Macro-average (AUC = {roc_auc["macro"]:.3f})',
        color='navy', linestyle=':', linewidth=3)

ax.plot([0, 1], [0, 1], 'k--', lw=2)
ax.set_xlim([0.0, 1.0])
ax.set_ylim([0.0, 1.05])
ax.set_xlabel('False Positive Rate', fontsize=12)
ax.set_ylabel('True Positive Rate', fontsize=12)
ax.set_title('Macro & Micro Average ROC Curves', fontsize=16)
ax.legend(loc="lower right", fontsize=12)
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

#Image Preprocessing for CNN

```python
# =============================
# 1. Import Libraries
# =============================
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
import torchvision.transforms as transforms
import torchvision.datasets as datasets
from torchvision.models import resnet50, ResNet50_Weights
import os
```

```python
# =============================
# 2. Setup Device
# =============================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
```

```python
# =============================
# 4. Image Transformations
# =============================
IMAGE_SIZE = 224

train_transform = transforms.Compose([
    transforms.Resize((IMAGE_SIZE, IMAGE_SIZE)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.ColorJitter(brightness=0.2, contrast=0.2),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])

val_transform = transforms.Compose([
    transforms.Resize((IMAGE_SIZE, IMAGE_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])
```

```python
# =============================
# 5. Create Datasets and DataLoaders
# =============================
BATCH_SIZE = 32

train_dataset = datasets.ImageFolder(TRAIN_DIR, transform=train_transform)
val_dataset = datasets.ImageFolder(VALID_DIR, transform=val_transform)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=2)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2)

num_classes = len(train_dataset.classes)
print(f"Classes: {train_dataset.classes}")
print(f"Number of classes: {num_classes}")
print(f"Train images: {len(train_dataset)}")
print(f"Validation images: {len(val_dataset)}")
```

```python
# =============================
# 6. Load Pretrained VGG16 Model
# =============================
import torchvision.models as models

model = models.vgg16(pretrained=True)

# Freeze convolutional layers
for param in model.features.parameters():
    param.requires_grad = False

# Replace classifier with correct number of classes
model.classifier[6] = nn.Linear(4096, num_classes)

model = model.to(device)

print("\n" + "="*60)
print("MODEL: VGG16 (Pretrained)")
print("="*60)
```

```python
# =============================
# 7. Loss & Optimizer
# =============================
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.classifier.parameters(), lr=0.0005)
```

```python
# =============================
# 8. Training Loop
# =============================
EPOCHS = 10

# History tracking
history = {
    'train_loss': [],
    'train_acc': [],
    'val_acc': []
}

print("\n" + "="*60)
print("STARTING CNN TRAINING")
print("="*60)

for epoch in range(EPOCHS):
    # Training phase
    model.train()
    total_loss = 0
    train_correct = 0
    train_total = 0

    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        preds = model(images)
        loss = criterion(preds, labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

        # Calculate training accuracy
        _, predicted = torch.max(preds, 1)
        train_total += labels.size(0)
        train_correct += (predicted == labels).sum().item()

    train_acc = train_correct / train_total * 100
    avg_loss = total_loss / len(train_loader)

    # Validation phase
    model.eval()
    val_correct = 0
    val_total = 0

    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)
            preds = model(images)
            _, predicted = torch.max(preds, 1)
            val_total += labels.size(0)
            val_correct += (predicted == labels).sum().item()

    val_acc = val_correct / val_total * 100

    # Store history
    history['train_loss'].append(avg_loss)
    history['train_acc'].append(train_acc)
    history['val_acc'].append(val_acc)

    print(f"Epoch {epoch+1}/{EPOCHS} | "
          f"Loss: {avg_loss:.4f} | "
          f"Train Acc: {train_acc:.2f}% | "
          f"Val Acc: {val_acc:.2f}%")

print("\n" + "="*60)
print("CNN TRAINING COMPLETED!")
print("="*60)
```

```python
# =============================
# 10. Get Predictions for Evaluation
# =============================
import numpy as np
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report, roc_curve, auc
from sklearn.preprocessing import label_binarize
import matplotlib.pyplot as plt
from itertools import cycle

print("\n" + "="*60)
print("GENERATING PREDICTIONS FOR EVALUATION")
print("="*60)

model.eval()
all_preds = []
all_labels = []
all_probs = []

with torch.no_grad():
    for images, labels in val_loader:
        images = images.to(device)
        outputs = model(images)
        probs = torch.softmax(outputs, dim=1)
        _, predicted = torch.max(outputs, 1)

        all_preds.extend(predicted.cpu().numpy())
        all_labels.extend(labels.numpy())
        all_probs.extend(probs.cpu().numpy())

all_preds = np.array(all_preds)
all_labels = np.array(all_labels)
all_probs = np.array(all_probs)
```

```python
# =============================
# 11. Confusion Matrix
# =============================
print("\n" + "="*60)
print("CONFUSION MATRIX")
print("="*60)

cm = confusion_matrix(all_labels, all_preds)
fig, ax = plt.subplots(figsize=(10, 8))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=train_dataset.classes)
disp.plot(cmap='Blues', ax=ax, xticks_rotation=45)
plt.title('Confusion Matrix - VGG16 CNN', fontsize=16, pad=20)
plt.tight_layout()
plt.show()
```

```python
# =============================
# 12. Classification Report
# =============================
print("\n" + "="*60)
print("CLASSIFICATION REPORT")
print("="*60)
report = classification_report(all_labels, all_preds, target_names=train_dataset.classes)
print(report)
```

## CUSTOM CNN MODEL (FROM SCRATCH)

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
import torchvision.transforms as transforms
import torchvision.datasets as datasets
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report, roc_curve, auc
from sklearn.preprocessing import label_binarize
from itertools import cycle
import re
```

```python
# =============================
# 1. Define Custom CNN Architecture
# =============================
class CustomCNN(nn.Module):
    def __init__(self, num_classes=10):
        super(CustomCNN, self).__init__()

        # Convolutional Block 1
        self.conv1 = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(inplace=True),
            nn.Conv2d(32, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )

        # Convolutional Block 2
        self.conv2 = nn.Sequential(
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )

        # Convolutional Block 3
        self.conv3 = nn.Sequential(
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.Conv2d(128, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )

        # Fully Connected Layers
        self.classifier = nn.Sequential(
            nn.Dropout(0.5),
            nn.Linear(128 * 28 * 28, 512),  # Adjusted for 224x224 input -> 28x28 after 3 pooling layers
            nn.ReLU(inplace=True),
            nn.Dropout(0.5),
            nn.Linear(512, 256),
            nn.ReLU(inplace=True),
            nn.Linear(256, num_classes)
        )

    def forward(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.conv3(x)
        x = x.view(x.size(0), -1)  # Flatten
        x = self.classifier(x)
        return x
```

```python
# =============================
# 2. Setup Device
# =============================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
```

```python
# =============================
# 3. Data Transformations
# =============================
IMAGE_SIZE = 224
BATCH_SIZE = 32

train_transform = transforms.Compose([
    transforms.Resize((IMAGE_SIZE, IMAGE_SIZE)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(15),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

val_transform = transforms.Compose([
    transforms.Resize((IMAGE_SIZE, IMAGE_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])
```

```python
# =============================
# 4. Create Datasets and DataLoaders
# =============================
train_dataset = datasets.ImageFolder(TRAIN_DIR, transform=train_transform)
val_dataset = datasets.ImageFolder(VALID_DIR, transform=val_transform)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=2)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2)

num_classes = len(train_dataset.classes)
print(f"\nClasses: {train_dataset.classes}")
print(f"Number of classes: {num_classes}")
print(f"Train images: {len(train_dataset)}")
print(f"Validation images: {len(val_dataset)}")
```

```python
# =============================
# 5. Initialize Custom CNN Model
# =============================
print("\n" + "="*60)
print("MODEL: Custom CNN (Built from Scratch)")
print("="*60)

model_custom = CustomCNN(num_classes=num_classes).to(device)

# Print model architecture
print(model_custom)

# Count parameters
total_params = sum(p.numel() for p in model_custom.parameters())
trainable_params = sum(p.numel() for p in model_custom.parameters() if p.requires_grad)
print(f"\nTotal Parameters: {total_params:,}")
print(f"Trainable Parameters: {trainable_params:,}")
```

```python
# =============================
# 6. Loss Function & Optimizer
# =============================
criterion_custom = nn.CrossEntropyLoss()
optimizer_custom = optim.Adam(model_custom.parameters(), lr=0.001)

# Learning rate scheduler
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer_custom, mode='max',
                                                   factor=0.5, patience=2)
```

```python
# =============================
# 7. Training Loop
# =============================
EPOCHS_CUSTOM = 15

history_custom = {
    'train_loss': [],
    'train_acc': [],
    'val_acc': []
}

print("\n" + "="*60)
print("STARTING CUSTOM CNN TRAINING")
print("="*60)

best_val_acc = 0.0

for epoch in range(EPOCHS_CUSTOM):
    # Training phase
    model_custom.train()
    total_loss = 0
    train_correct = 0
    train_total = 0

    for batch_idx, (images, labels) in enumerate(train_loader):
        images, labels = images.to(device), labels.to(device)

        optimizer_custom.zero_grad()
        outputs = model_custom(images)
        loss = criterion_custom(outputs, labels)
        loss.backward()
        optimizer_custom.step()

        total_loss += loss.item()
        _, predicted = torch.max(outputs, 1)
        train_total += labels.size(0)
        train_correct += (predicted == labels).sum().item()

    train_acc = train_correct / train_total * 100
    avg_loss = total_loss / len(train_loader)

    # Validation phase
    model_custom.eval()
    val_correct = 0
    val_total = 0

    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model_custom(images)
            _, predicted = torch.max(outputs, 1)
            val_total += labels.size(0)
            val_correct += (predicted == labels).sum().item()

    val_acc = val_correct / val_total * 100

    # Update learning rate
    scheduler.step(val_acc)

    # Store history
    history_custom['train_loss'].append(avg_loss)
    history_custom['train_acc'].append(train_acc)
    history_custom['val_acc'].append(val_acc)

    # Save best model
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model_custom.state_dict(), 'best_custom_cnn.pth')

    print(f"Epoch {epoch+1}/{EPOCHS_CUSTOM} | "
          f"Loss: {avg_loss:.4f} | "
          f"Train Acc: {train_acc:.2f}% | "
          f"Val Acc: {val_acc:.2f}%")

print("\n" + "="*60)
print("CUSTOM CNN TRAINING COMPLETED!")
print(f"Best Validation Accuracy: {best_val_acc:.2f}%")
print("="*60)
```

```python
# =============================
# 9. Generate Predictions
# =============================
print("\n" + "="*60)
print("GENERATING PREDICTIONS FOR CUSTOM CNN")
print("="*60)

all_preds_custom = []
all_labels_custom = []
all_probs_custom = []

with torch.no_grad():
    for images, labels in val_loader:
        images = images.to(device)
        outputs = model_custom(images)
        probs = torch.softmax(outputs, dim=1)
        _, predicted = torch.max(outputs, 1)

        all_preds_custom.extend(predicted.cpu().numpy())
        all_labels_custom.extend(labels.numpy())
        all_probs_custom.extend(probs.cpu().numpy())

all_preds_custom = np.array(all_preds_custom)
all_labels_custom = np.array(all_labels_custom)
all_probs_custom = np.array(all_probs_custom)
```

```python
# =============================
# 10. Confusion Matrix
# =============================
print("\n" + "="*60)
print("CONFUSION MATRIX - CUSTOM CNN")
print("="*60)

cm_custom = confusion_matrix(all_labels_custom, all_preds_custom)
fig, ax = plt.subplots(figsize=(10, 8))
disp = ConfusionMatrixDisplay(confusion_matrix=cm_custom, display_labels=train_dataset.classes)
disp.plot(cmap='Purples', ax=ax, xticks_rotation=45)
plt.title('Confusion Matrix - Custom CNN', fontsize=16, pad=20)
plt.tight_layout()
plt.show()
```

```python
# =============================
# 12. Accuracy & Loss Curves
# =============================
print("\n" + "="*60)
print("ACCURACY & LOSS CURVES - CUSTOM CNN")
print("="*60)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

epochs_range = range(1, EPOCHS_CUSTOM + 1)

# Loss curve
ax1.plot(epochs_range, history_custom['train_loss'], 'purple', label='Training Loss', linewidth=2)
ax1.set_xlabel('Epoch', fontsize=12)
ax1.set_ylabel('Loss', fontsize=12)
ax1.set_title('Custom CNN Training Loss', fontsize=14)
ax1.legend(fontsize=10)
ax1.grid(alpha=0.3)

# Accuracy curves
ax2.plot(epochs_range, history_custom['train_acc'], 'purple', label='Training Accuracy', linewidth=2)
ax2.plot(epochs_range, history_custom['val_acc'], 'darkorange', label='Validation Accuracy', linewidth=2)
ax2.set_xlabel('Epoch', fontsize=12)
ax2.set_ylabel('Accuracy (%)', fontsize=12)
ax2.set_title('Custom CNN Training & Validation Accuracy', fontsize=14)
ax2.legend(fontsize=10)
ax2.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python

# =============================
# 13. Extract Custom CNN Metrics
# =============================
print("\n" + "="*60)
print("EXTRACTING CUSTOM CNN METRICS")
print("="*60)

# Overall accuracy
custom_overall_accuracy = (all_preds_custom == all_labels_custom).sum() / len(all_labels_custom)

# ROC AUC scores
n_classes = len(train_dataset.classes)
y_val_bin_custom = np.eye(n_classes)[all_labels_custom]

fpr_custom = dict()
tpr_custom = dict()
roc_auc_custom = dict()

for i in range(n_classes):
    fpr_custom[i], tpr_custom[i], _ = roc_curve(y_val_bin_custom[:, i], all_probs_custom[:, i])
    roc_auc_custom[i] = auc(fpr_custom[i], tpr_custom[i])

# Store metrics
custom_metrics = {
    'overall_accuracy': custom_overall_accuracy,
    'class_metrics': {}
}

custom_report_lines = report_custom.split('\n')
class_names = train_dataset.classes

for line in custom_report_lines:
    match = re.match(r'^\s*(\S+)\s+([0-9.]+)\s+([0-9.]+)\s+([0-9.]+)\s+([0-9]+)$', line)
    if match:
        cls = match.group(1)
        if cls in class_names:
            custom_metrics['class_metrics'][cls] = {
                'precision': float(match.group(2)),
                'recall': float(match.group(3)),
                'f1-score': float(match.group(4)),
                'support': int(match.group(5))
            }

for i, class_name in enumerate(class_names):
    if class_name in custom_metrics['class_metrics']:
        custom_metrics['class_metrics'][class_name]['auc'] = roc_auc_custom[i]

print("Custom CNN Metrics:")
for key, value in custom_metrics.items():
    if isinstance(value, dict):
        print(f"  {key}:")
        for sub_key, sub_value in value.items():
            print(f"    {sub_key}: {sub_value}")
    else:
        print(f"  {key}: {value}")
```
